### Script de Extração


```python
def extrair_dados():
    print("Extraindo dados...")
    return [
        {"cliente": "Ana", "valor": 5000},
        {"cliente": "Carlos", "valor": 100},
        {"cliente": "Ana", "valor": 300}
    ]
```

### Script de Transformação

```python
def transformar_dados(dados):
    print("Transformando dados...")
    resultado = {}

    for d in dados:
        cliente = d["cliente"]
        valor = d["valor"] * 1.1

        if cliente not in resultado:
            resultado[cliente] = 0

        resultado[cliente] += valor

    return resultado
```

### Script de Carga


```python
def carregar_dados(dados):
    print("Carregando dados...")
    for cliente, total in dados.items():
        print(f"Inserindo: {cliente} - {total}")
```

### Orquestrador

```python
def pipeline_orquestrador():
    dados = extrair_dados()
    dados_transformados = transformar_dados(dados)
    carregar_dados(dados_transformados)

```

```python
pipeline_orquestrador()
```
